# Этап 6

Исходная система уравнений:
$$\left\{ \begin{array} a\dot{x}=x+y\\ \dot{y}=-7x+3y \end{array} \right. \: x(0)=6, y(0)=4$$
корни характеристического уравнения:
$$\lambda_{1,2} = 2\pm\sqrt{6}i$$

Так как корни характеристического уравнения являются комплексными числами и $Re\lambda_1 = Re\lambda_2 = 2 >0$, то точка покоя (0, 0) представляет собой **неустойчивый фокус**

Решения системы:
$$
\begin{array}
    xx = e^{2t}\left(C_1 cos{\sqrt{6}t} +C_2 sin{\sqrt{6}t}\right)\\
    y = e^{2t}\left((C_1+\sqrt{6} C_2)cos{\sqrt{6}t} + (C_2-\sqrt{6} C_1) sin{\sqrt{6}t}\right)
\end{array}
$$
А с учетом начальных условий:
$$
\begin{array}
    ax = e^{2t}\left(6 cos{\sqrt{6}t} - \frac{2}{\sqrt{6}} sin{\sqrt{6}t}\right)\\
    y = e^{2t}\left(4 cos{\sqrt{6}t} + \left(- \frac{2}{\sqrt{6}} - 6\sqrt{6} \right) sin{\sqrt{6}t}\right)
\end{array}
$$

Для выяснения вопроса о направлении «закручивания» траекторий вычислим правую часть системы при x = 4, y = 0 — получим $\dot x = 4, \dot y = -28$, траектория раскручивается **по часовой стрелке**

Запишем уравнение изоклин:
$$\frac{dx}{dy}=k=\frac{x+y}{-7x+3y}$$

Построим таблицу для различных k:

|Значение k|Ур-ие изоклины|
|:-:|:-:|
|$-1/4$|$y=\frac{3}{7} x$|
|$-1/7$|$y=0$|
|$0$|$y=-x$|
|$1/5$|$y=-6x$|
|$1$|$y=4x$|

```python
x=np.arange(0,3,1)

k=-1/4
y=k*x

plt.figure(figsize=(4, 15))

plt.subplot(511)
plt.plot(x,y)
plt.gca().set_title('k=-1/4')
plt.axis('equal')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.grid()


k=-1/7
y=k*x
plt.subplot(512)
plt.plot(x,y)
plt.gca().set_title('k=-1/7')
plt.axis('equal')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.grid()

k=0
y=k*x
plt.subplot(513)
plt.plot(x,y)
plt.gca().set_title('k=0')
plt.axis('equal')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.grid()

k=1/5
y=k*x
plt.subplot(514)
plt.plot(x,y)
plt.gca().set_title('k=1/5')
plt.axis('equal')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.grid()

k=1
y=k*x
plt.subplot(515)
plt.plot(x,y)
plt.gca().set_title('k=1')
plt.axis('equal')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.grid()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

for sk, k, slope, text_x, text_y in zip(
            ['-1/4', '-1/7', '0', '1/5', '1'],
            [-1/4, -1/7, 0, 1/5, 1],
            [3/7, 0, -1, -6, 4],
            [18, 18, -20, -3, 5],
            [5, -3, 21, 25, 15],
        ):
    x = np.arange(-30,30,5)
    b, e = x-1, x+1
    y = slope*x
    plt.plot(x, y, '-b')
    plt.text(text_x, text_y, 'k={}'.format(sk))
    if k==0:
        plt.plot(np.stack((x,x)), np.stack([2+y, -2+y]), 'r', linewidth=1)
    else:
        for xb, xe, yy in zip(b, e, y):
            plt.plot([xb, xe], [-1/k+yy, 1/k+yy], 'r')


C1 = [6, -2, 2, -4]
C2 = [-2/np.sqrt(6), 2, -2, -4]
colors = ['red', 'black', 'black', 'black']

t = np.arange(-10,2,0.05)
for c1, c2, color in zip(C1, C2, colors):
    x = np.exp(2*t)*(c1*np.cos(np.sqrt(6)*t)+c2*np.sin(np.sqrt(6)*t))
    y = np.exp(2*t)*((c1+np.sqrt(6)*c2)*np.cos(np.sqrt(6)*t)+(c2-np.sqrt(6)*c1)*np.sin(np.sqrt(6)*t))

    plt.plot(x, y, color=color, linewidth=2)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.xlim(-30, 30)
plt.ylim(-30, 30)

plt.arrow(4, 0, 4, -28, head_width=2, head_length=2, fc='orange', ec='orange')
# plt.plot(x[0], y[0], 'yo')
```
