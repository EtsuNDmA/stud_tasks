```python
# Moving average
import numpy as np
def moving_average(x, N):
    result = np.zeros([len(x), len(N)])
    cumsum = np.cumsum(np.insert(tuple(x), 0, 0))
    for i, n in enumerate(N):
        result[n-1:,i] = (cumsum[n:] - cumsum[:-n])/float(n)
    return result

N = 10, 20, 30, 40, 50
names_ma = ['MovingAVR1', 'MovingAVR2', 'MovingAVR3', 'MovingAVR4', 'MovingAVR5']
data_ma = pd.DataFrame(moving_average(data['averageFB'], N), columns=names_ma)
```
